# Import libary

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
```

<h1 align='center'>PCA for Wine Classification</h1>

##Bước 1: Đọc file Wine.csv

```python
link_data = '/gdrive/MyDrive/Colab Notebooks/Lập trình Python cho Máy học/Week 5/Wine.csv'
```

```python
data_wine = pd.read_csv(link_data)
data_wine.head()
```

```python
df_wine = pd.DataFrame(data_wine)
X_data = df_wine.iloc[:, 0:-1];
Y_data = df_wine.iloc[:, -1:];
```

```python
X_data
```

```python
Y_data
```

##Bước 2: Chuẩn hóa đặc trưng đầu vào X bằng StandardScaler

```python
X_data_np = X_data.to_numpy()
Y_data_np = Y_data.to_numpy()
std_X_np = StandardScaler().fit_transform(X_data_np)
std_X_df = pd.DataFrame(std_X_np)
std_X_df.head()
```

##Bước 3: PCA để nén đặc trưng đầu vào từ 13 chiều xuống còn 2 chiều

```python
from sklearn.decomposition import PCA
pca = PCA(2)
pca_std_X_np = pca.fit_transform(std_X_np)
pca_std_X_np
```

##Bước 4: Chia tập dữ liệu của mình ra thành 2 phần train:test = 8:2

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(pca_std_X_np, Y_data.to_numpy(), test_size=0.2)
```

##Bước 5: Dùng Logistic Regeression để phân lớp dữ liệu

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

##Bước 6: Đánh giá mô hình với độ đo accuracy

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
```

```python
print(classification_report(Y_test, Y_pred))
```

#Nhận xét

- Mỗi lần split dữ liệu sẽ nhận được một bộ dữ liệu khác nhau, nên độ chính xác thu được sau mỗi lần train không ổn định

##Bước 7: Đánh model giá bằng K-Fold Cross Validation



```python
from sklearn.model_selection import KFold
```

```python
kfold = KFold(5, True, 1)
```

```python
Y_data_np.shape
```

```python
accuracy_list = []
for train, test in kfold.split(X_data_np, Y_data_np):
    X_train, X_test = X_data_np[train], X_data_np[test];
    Y_train, Y_test = Y_data_np[train], Y_data_np[test]

    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    accuracy_list.append(accuracy_score(Y_test, Y_pred))
```

```python
accuracy_list
```

```python
np.mean(accuracy_list)
```
