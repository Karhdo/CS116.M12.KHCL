```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Tạo dữ liệu hình sin
t = np.arange(0.0, 2.0, 0.05) #t Lấy mẫu từ 0 đến 2, bước nhảy 0.05
s = np.sin(2 * np.pi * t) # s tính theo t: s = sin(2*pi*t)
```

```python
#Vẽ dạng đường và điểm
plt.plot(t, s)
plt.plot(t, s+1, 'r^')
plt.plot(t, s-1, 'go')

#Đặt tên đồ thị, Ox, Oy
plt.title('Biểu đồ dạng sóng hình sin', fontsize=15)
plt.xlabel('Trục thời gian (t)')
plt.ylabel('Trục biên độ (s)')

#Vẽ text dựa vào toạ độ
plt.text(1.75, -0.3, r'$s=\mathrm{sin}(2 \pi t)$')
plt.text(1.75, 0.8, r'$s=\mathrm{sin}(2 \pi t + 1)$')
plt.text(1.75, -1.2, r'$s=\mathrm{sin}(2 \pi t - 1)$')

#Giới hạn của trục Ox, Oy
plt.xlim(-0.5, 2.5)
plt.ylim(-2.5, 2.5)
```

####Vẽ nhiều biểu đồ với hàm: subplot (<nrow><ncol><index>)

```python
# Vẽ ô thứ 1 trong bảng 2x3 
# Vẽ biểu đồ dạng điểm
plt.subplot(231)
plt.scatter(t, s, marker=">")

# Vẽ ô thứ 2 trong bảng 2x3 
# Vẽ biểu đồ dạng đường
plt.subplot(232)
plt.plot(t, s)

# Vẽ ô thứ 3 trong bảng 2x3 
# Vẽ biểu đồ cột
plt.subplot(233)
x = np.arange(3)
money = [1.5e5, 2.5e6, 5.5e6]
plt.bar(x, money)
plt.xticks(x, ('Cty A', 'Cty B', 'Cty C'))

# Vẽ ô thứ 4 trong bảng 2x3
# Vẽ biểu đồ dạng tròn
plt.subplot(234)
labels = 'Others', 'China', 'USA'
sizes = [23, 35, 40]
plt.pie(sizes, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)

# Vẽ ô thứ 5 trong bảng 2x3
# Vẽ biểu đồ xếp chồng
plt.subplot(235)
x = [1,2,3,4,5]
y1 = [1,1,2,3,5]
y2 = [0,4,2,6,8]
y3 = [1,3,5,7,9]
y = np.vstack([y1, y2, y3])
plt.stackplot(x,y1,y2,y3, labels=labels)

# Vẽ ô thứ 5 trong bảng 2x3
# Vẽ biểu đồ xếp chồng
plt.subplot(236)
image = plt.imread('/content/gdrive/MyDrive/Colab Notebooks/Computer Vision/jurassic-park.png')
plt.imshow(image)
plt.axis('off')
```

##Nối dữ liệu theo chiều dọc với concat

```python
# Nối hai data frame theo chiều dọc
df1 = pd.DataFrame({
        "X" : ['A', 'B', 'O', 'AB'],
        "Y" : [15, 12, 10, 12],
        "Z" : [30, 28, 23, 29]},
        index = [1, 2, 3, 4])

df2 = pd.DataFrame({
        "X" : ['O', 'A', 'B'],
        "Y" : [20, 21, 22],
        "Z" : [32, 30, 20],
        "T" : [1, 0, 1]},
        index = [1, 2, 3])

df_new = pd.concat([df1, df2])
df_new
```

```python
#Điền giá trị khuyết với fillna(value)
df_new.fillna(-1)
```

```python
#Nối dữ liệu theo chiều ngang với concat (axis=1)
df1 = pd.DataFrame({
    "X" : ['A', 'B', 'O', 'AB'],
    "Y" : [15, 12, 10, 12],
    "Z" : [30, 28, 23, 29]},
    index = [1, 2, 3, 4])

df2 = pd.DataFrame({
    "U" : [0, 1, 0],
    "V" : [20, 1, 6]},
    index = [1, 2, 3])

pd.concat([df1, df2], axis=1)
```
