##Data Manipulation

```python
df.item_price.dtype #Trả về kiểu dữ liệu của cột item_price
```

##Apply()

```python
df.item_price.apply(lambda x : float(x.replace('$', ''))).dtype
```

```python
df.item_price = df.item_price.apply(lambda x : float(x.replace('$', '')))
```

```python
df.head()
```

## Tạo và nhân các columns trong Data

```python
df['total_price'] = df['quantity']*df['item_price'] #Tạo một cột mới total_price
# or df.total_price = df.quantity * df.item_price
```

```python
df.head()
```

###How much was the revenue for the period in the dataset?

```python
revenue = df.total_price.sum()
print(revenue)
```

##Group By

###Which was the most-odered item?

```python
quantity_item = df.groupby('item_name')['quantity'].sum() #Gom nhóm theo item_name và sum số lượng
```

```python
quantity_item
```

```python
quantity_item.sort_values(ascending=False).head(5)
```
