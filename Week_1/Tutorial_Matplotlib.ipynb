#Pyplot API vs Oriented API
- Pyplot API -> Quicky
- Oriented API -> Advanced

##Pyplot API -> Quicky

```python
x = np.linspace(0, 10, 1000) #x từ 0 -> 10 và gồm 1000 dữ liệu x
plt.plot(x, np.sin(x), color='blue', linestyle='dashed', label='sin(x)');
plt.plot(x, np.cos(x), color='red', label='cos(x)');

plt.title('sin(x) and cos(x)')
plt.xlabel('x'); #Label cho trục Ox, Oy
plt.ylabel('F(x)');

#Giới hạn độ dài của trục Ox và Oy
#plt.xlim([0, 8]) 
#plt.ylim([-1, 0.75])
#plt.axis(xmin, xmax, ymin, ymax)
plt.axis("equal")

plt.legend(); #Dùng để hiện label trên đồ thị.
```

##Oriented API

```python
#1 Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2 Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = width and height of the plot

#3 Plot data
ax.plot(x, y);
ax.set(title='Simple plot',
       xlabel='x-axis',
       ylabel='y-axis');
```

```python
#OO API

#Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
size = 1000*rng.rand(100)

fig, ax = plt.subplots()

plot_1 = ax.scatter(x, y, s=size, c=colors, cmap='viridis', alpha=0.3); #cmap: chỉnh theme màu, alpha: chỉnh độ trong suốt của màu

fig.colorbar(plot_1);
```

##Bar chart
- Vertical
- Horizontal

```python
# Prepare Data
soft_drink_prices = {'CoaCoLa':10, 'Pesi': 12, 'Sprite': 8} 

fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())

ax.set(title='Soft drink prices', xlabel='Name', ylabel='Prices ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))
```

##Histogram

```python
#Prepare data
np.random.seed(42) #Dùng để random lần nào cũng ra giống nhau
student_height = np.random.normal(170, 10, 250) # (giá trị trung bình, độ lệch chuẩn, số lượng)
student_height[:10]
```

```python
#pylot API
plt.hist(student_height, bins=30); #bins dùng để chia nhỏ dữ liệu
```

#Subplots

```python
# Option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins=30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
#Plot data to each axis
ax[0,0].plot(x, x/2);
ax[0,0].plot(x, x*2);

ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins=30);
```
